```python
# Removing rows which are null
df1=df.dropna()
df1.head()
```

```python
# Checking if their is any null value left
df1.isna().sum()
```

```python
# Converting the dataset to dataframe to make plotting easier
data=pd.DataFrame(df1)
x = data['Region'].value_counts()
```

```python
# Plotting the Distribution of Countries by their Region
x.plot(kind='bar', color='blue')
plt.xlabel('Region')
plt.ylabel('No of Countries')
plt.title('Distribution of Countries by their Region')
plt.xticks(rotation=45) 
plt.show()
```

```python
y = data['Death rate, crude (per 1,000 people)'] 
```

```python
# Plotting of Frequency of Death rate, crude (per 1000p eople)
y.plot(kind='hist',color='orange',edgecolor='black')
plt.title('Frequency of Death rate, crude (per 1,000 people)')
plt.xlabel('Death rate, crude (per 1,000 people)')
```
